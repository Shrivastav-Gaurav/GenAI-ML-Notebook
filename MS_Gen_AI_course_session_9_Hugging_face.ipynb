```python
# --- Install Hugging Face Diffusers & dependencies ---
%pip install diffusers transformers accelerate safetensors pillow

# --- Imports ---
import torch
from diffusers import StableDiffusionPipeline, StableDiffusionImg2ImgPipeline
from PIL import Image
import requests
import IPython.display as display
from google.colab import files

# --- Device selection (use GPU if available) ---
device = "cuda" if torch.cuda.is_available() else "cpu"

# --- Load pipelines ---
# Text-to-Image
pipe_txt2img = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16
).to(device)

# Image-to-Image
pipe_img2img = StableDiffusionImg2ImgPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16
).to(device)

def generate_image(prompt, input_image_path=None, strength=0.75, guidance_scale=7.5):
    """
    prompt: text description
    input_image_path: path to existing image (optional)
    strength: how much to transform image (0.0 = no change, 1.0 = fully new)
    guidance_scale: higher = more prompt adherence
    """
    if input_image_path:
        # Open input image
        init_image = Image.open(input_image_path).convert("RGB").resize((512, 512))

        # Run Image-to-Image
        result = pipe_img2img(
            prompt=prompt,
            image=init_image,
            strength=strength,
            guidance_scale=guidance_scale
        ).images[0]

        # Show both images
        display.display(display.HTML("<h3>Original Image</h3>"))
        display.display(init_image)
        display.display(display.HTML("<h3>Generated Image</h3>"))
        display.display(result)

    else:
        # Run Text-to-Image
        image = pipe_txt2img(
            prompt=prompt,
            guidance_scale=guidance_scale
        ).images[0]

        display.display(display.HTML("<h3>Generated Image</h3>"))
        display.display(image)

# --- Example usage ---
# OPTION 1: Just from prompt
generate_image("A futuristic city skyline at sunset, cyberpunk style")

# OPTION 2: Upload an image first, then run
# uploaded = files.upload()
# img_path = list(uploaded.keys())[0]
# generate_image("Add a rainbow in the sky", input_image_path=img_path)
```
